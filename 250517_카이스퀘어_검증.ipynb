```python
 from google.colab import files
myfile = files.upload()
```

```python
import pandas as pd
import io
```

```python
df = pd.read_csv(io.BytesIO(myfile['data_chi.csv']))
```

```python
df
```

```python
table = pd.crosstab(df['Diabetes'], df['Obesity'])
```

```python
print(table)
```

```python
import seaborn as sns
sns.heatmap(table, annot=True, cmap = 'Blues')
```

```python
from scipy.stats import chi2_contingency
```

```python
chi2_contingency(table)
```

```python
chi2, p, dof, expected = chi2_contingency(table)
```

```python
if p < 0.05:
  print('두 변수는 독립이 아니다 = 두 변수는 관련성이 있다.')
else:
  print('두 변수는 독립이다 = 두 변수는 관련성이 없다')
```
