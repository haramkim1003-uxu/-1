```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier()
```

```python
kn.fit(fruit_data, fruit_target)
```

```python
kn.predict([[8,140]])
```

```python
kn.score(fruit_data, fruit_target)
```
