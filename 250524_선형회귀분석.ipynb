```python
from google.colab import files
myfile = files.upload()
```

```python
import pandas as pd
import io
```

```python
df = pd.read_csv(io.BytesIO(myfile['heights.csv']))
```

```python
df
```

```python
from pylab import plot
```

```python
x = df['height']
y = df['weight']
```

```python
plot(x, y, 'o')
```

```python
#Scikit-learn = Science + kit + learn
from sklearn.linear_model import LinearRegression
```

```python
line_fitter = LinearRegression()
```

```python
x.values.reshape(-1,1)
```

```python
line_fitter.fit(x.values.reshape(-1,1), y)
```

```python
line_fitter.coef_
```

```python
line_fitter.intercept_
```

```python
#y = ax + b
#y = 3.43x - 106.02
```

```python
line_fitter.predict([[73]])
```

```python
plot(x,y,'o',x,line_fitter.predict(x.values.reshape(-1,1)), 'r-')
```
