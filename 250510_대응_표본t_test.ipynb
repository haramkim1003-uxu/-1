상황: 어떤반의 10 명의 시헝성적을 비교

 *3월:70,80,85,75,80,85,95,85,80,80

 *8월80,85,80,90,95,80,100,90,95,90

```python
march = [70,80,85,75,80,85,95,85,80,80]
august = [80,85,80,90,95,80,100,90,95,90]
```

2.박스 플롯으로 시각화하여 나타내기

```python
import matplotlib.pyplot as plt
```

```python
plt.boxplot([march, august])
plt.xlabel('when')
plt.ylabel('score')
plt.grid()
```

3.정규성 검사(shapiro-Wilks test)

```python
from scipy.stats import shapiro
```

```python
shapiro(march)
```

```python
shapiro(august)
```

4.대응표본 t-test 수행하기(정규성을 만족했기 때문)

```python
from scipy.stats import ttest_rel
```

```python
ttest_rel(march, august)
```

```python
import numpy as np
```

```python
np.mean(march)
```

```python
np.mean(august)
```

만약, 정규성이 만족되지 못했다면 -> Wilcoxon Sign-ranked test(윌콕슨 부호화 테스트)

```python
from scipy.stats import wilcoxon
```

```python
wilcoxon(march, august)
```
