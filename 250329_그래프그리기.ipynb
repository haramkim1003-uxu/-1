Quiz. 2015년부터 2024년까지의 서울의 연도별 강수량을 세로형 막대그래프로 그리기

*   연도별 강수량 자료: 기상청 기상자료개방포털(https://data.kma.go.kr/stcs/grnd/grndRnList.do?pgmNo=69)
*   조건1: 막대그래프의 테두리를 그리기
*   조건2: 막대그래프의 색깔을 녹색으로 하기
*   조건3: y축 값의 범위를 log를 이용하여 10의 지수로 나타내기

image.png




```python
import matplotlib.pyplot as plt

years = [a for a in range(2015,2025)]
values = [792.1, 991.7, 1233.2, 1284.1, 891.3, 1651.1, 1186.5, 1775.3, 1598.8, 1311.4]

plt.bar(years, values, edgecolor='black', color='green', log=True)
```
