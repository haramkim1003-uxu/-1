```python
g1 = [90,95,80,100,95,70,90,95,100,85]
g2 = [80,95,100,85,90,90,95,80,65,95]
```

```python
import numpy as np
import scipy.stats

def ftest(x, y):
  x = np.array(x)
  y = np.array(y)

  f = np.var(x) / np.var(y)

  dfn = x.size - 1
  dfd = y.size - 1

  p = 1 - scipy.stats.f.cdf(f, dfn, dfd)

  return f,p
```

```python
#p값이 0.05 보다 작으면 두집단의 분산은 유의미한 차이가 있다.
ftest(g1,g2)
```
