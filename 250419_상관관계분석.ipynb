```python
height = [170,168,177,181,172,171,169,175,174,178,170,167,177,182,173,171,170,179,175,177,186,166,183,168]
weight = [70,66,73,77,74,73,69,79,77,80,74,68,71,76,78,72,68,79,77,81,84,73,78,69]
```

```python
import pandas as pd
```

```python
body = pd.DataFrame({'height': height, 'weight':weight})
```

```python
body
```

```python
# @title height vs weight

from matplotlib import pyplot as plt
body.plot(kind='scatter', x='height', y='weight', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#산점도(Scatter Plot)그리기
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(body['height'],body['weight'])
plt.xlabel('height')
plt.ylabel('weight')
plt.title('Height vs Weight')
```

```python
#상관관계 분석(Pearson의 상관관계분석)
body.corr(method = 'pearson')
```

Pearson의 상관관계 분석하기
*   0.8이상: 강한 상관
*   0.6~0.8: 보통의 상관
*   0.4~0.6: 약한 상관
*   0.4미만: 거의 상관없음
